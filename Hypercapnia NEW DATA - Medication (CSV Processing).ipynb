# Medication Preprocessing (NEW DATA)

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from datetime import datetime
import os
import glob

database_dir = "D:\\TriNetX\\Medications\\"  # Location where the database files are stored 
#working_dir = "D:\\TriNetX\\Medications\\"
working_dir = r"C:\Users\reblo\Box\Residency Personal Files\Scholarly Work\Locke Research Projects\TriNetX Code\Hypercapnia TriNetX CSV Processing\Working\\" #location where to read and right from (faster = better if space allows)
print(working_dir)
#num_spreadsheets = 50
num_spreadsheets = 2991
```

# Reformatting of CSVs

```python
# split -l 100000000 medication.csv
# Adding new indexes; adding new column names; removing un-needed columns; reformatting column data types

columns = ["patient_id","encounter_id","unique_id","code_system","code","start_date","route","brand","strength","derived_by_TriNetX","source_id"]

for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    medication = pd.read_csv(database_dir + "medication"+f'{i:04}'+".csv",
                            names = columns,
                            dtype = {"patient_id":str,"encounter_id":str,"unique_id":str,"code_system":str,"code":str,"route":str,"brand":str,"strength":str,"derived_by_TriNetX":str,"source_id":str},
                            parse_dates = ["start_date"])
    medication = medication.drop(columns = ["unique_id","code_system","route","brand","strength","derived_by_TriNetX","source_id"])
    medication.to_csv(working_dir + "medication_NEW_"+f'{i:04}'+".csv", index = False)
    print(medication.shape)
    del medication
```
