### Reformatting of CSVs

```python
# Adding new indexes; adding new column names; removing unneeded columns; reformatting column data types

columns = ["patient_id","encounter_id","code_system","code","principal_diagnosis_indicator","admitting_diagnosis","reason_for_visit","date","derived_by_TriNetX","source_id"]

for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    diagnosis = pd.read_csv(database_dir + "diagnosis"+f'{i:04}'+".csv",
                            names = columns,
                            dtype = {"patient_id":str,"encounter_id":str,"code":str},
                            parse_dates = ["date"])
    diagnosis.loc[diagnosis["principal_diagnosis_indicator"] == "Unknown", "principal_diagnosis_indicator"] = "U"  # Unknown, S U 
    diagnosis.loc[diagnosis["admitting_diagnosis"] == "Unknown", "admitting_diagnosis"] = "U" # F and U
    diagnosis.loc[diagnosis["reason_for_visit"] == "Unknown", "reason_for_visit"] = "U" # F and U
    diagnosis = diagnosis.drop(columns = ["code_system","derived_by_TriNetX","source_id"])
    diagnosis.to_csv(working_dir + "diagnosis_NEW_"+f'{i:04}'+".csv", index = False)
    print(diagnosis.shape)
    del diagnosis

#Test to ensure diagnosis indicators work
"""
for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    test = pd.read_csv(working_dir + "diagnosis_NEW_"+f'{i:04}'+".csv")
    #display(test.head(10))

    print(test["principal_diagnosis_indicator"].value_counts())
    print(test["admitting_diagnosis"].value_counts())
    print(test["reason_for_visit"].value_counts())
    test.loc[test["principal_diagnosis_indicator"] == "Unknown", "principal_diagnosis_indicator"] = "U"
    test.loc[test["admitting_diagnosis"] == "Unknown", "admitting_diagnosis"] = "U"
    test.loc[test["reason_for_visit"] == "Unknown", "reason_for_visit"] = "U"

    display(test.head(10))
    print(test["principal_diagnosis_indicator"].value_counts())
    print(test["admitting_diagnosis"].value_counts())
    print(test["reason_for_visit"].value_counts())
"""
```
