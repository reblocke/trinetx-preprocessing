# Lab Results Preprocessing (NEW DATA)

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from datetime import datetime
import os
import glob

database_dir = r"D:\TriNetX\Lab Results\\"   # Location where the database files are stored 
#working_dir = r"D:\TriNetX\Lab Results\"
working_dir = r"C:\Users\reblo\Box\Residency Personal Files\Scholarly Work\Locke Research Projects\TriNetX Code\Hypercapnia TriNetX CSV Processing\Working\\" #location where to read and right from (faster = better if space allows)
print(working_dir)
#num_spreadsheets = 10
num_spreadsheets = 2334
d_type_spec = {"patient_id":str, "encounter_id":str, "code":'category'}
```

### Reformatting of CSVs

```python
# Adding new indexes; adding new column names; removing un-needed columns; reformatting column data types

columns = ["patient_id","encounter_id","code_system","code","date","lab_result_num_val","lab_result_text_val","units_of_measure","derived_by_TriNetX","source_id"]
for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    lab_results = pd.read_csv(database_dir + "lab_results"+f'{i:04}'+".csv",
                            names = columns,
                            dtype = {"patient_id":str,"encounter_id":str,"code":str},
                            parse_dates = ["date"])
    lab_results = lab_results.drop(columns = ["code_system","lab_result_text_val","units_of_measure","derived_by_TriNetX","source_id"])
    lab_results.to_csv(working_dir + "lab_results_NEW_"+f'{i:04}'+".csv", index = False)
    print(lab_results.shape)
    del lab_results
```

# Potassium [Moles/volume] in Serum, Plasma or Blood; Processing
### LOINC: 6298-4 - Potassium [Moles/volume] in Blood
### LOINC: 2823-3 - Potassium [Moles/volume] in Serum or Plasma
### LOINC: 77142-8 - Potassium [Moles/volume] in Serum, Plasma or Blood

```python
# 6298-4: Potassium [Moles/volume] in Blood
# 2823-3: Potassium [Moles/volume] in Serum or Plasma
# 77142-8: Potassium [Moles/volume] in Serum, Plasma or Blood

sub_blocks = []
for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    lab_results = pd.read_csv(working_dir + "lab_results_NEW_"+f'{i:04}'+".csv", dtype = d_type_spec)
    lab_results= lab_results[lab_results["code"].str.match("^6298-4$|^2823-3$|^77142-8$")]
    lab_results["lab_result_num_val"] = lab_results["lab_result_num_val"].astype('float16') # to save memory; shouldn't ahve any values outside 32k
    lab_results = lab_results.loc[(lab_results["lab_result_num_val"] < 12) & (lab_results["lab_result_num_val"] >= 1.8)]
    sub_blocks.append(lab_results)
    del lab_results

value_potassium = pd.concat(sub_blocks, ignore_index = True)
value_potassium.to_csv(working_dir +"value_potassium.csv", index = False)
display(value_potassium.head())
print("\n")
print(value_potassium.shape)
print("\n")
print(value_potassium["code"].value_counts())

##### Histogram of value_potassium #####
plt.figure(figsize = (10,10))
plt.hist(value_potassium["lab_result_num_val"], bins = 100, facecolor = "blue")
plt.title("Histogram of value_potassium", color = "black", fontsize = 12)
plt.xlabel("Value", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.show()

del value_potassium
```
