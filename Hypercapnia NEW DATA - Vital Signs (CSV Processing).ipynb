# Vital Signs Preprocessing (NEW DATA)

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import psutil
import os
import glob

database_dir = r"D:\TriNetX\Vital Signs\\"   # Location where the database files are stored 
#working_dir = r"D:\TriNetX\Vital Signs\"
working_dir = r"C:\Users\reblo\Box\Residency Personal Files\Scholarly Work\Locke Research Projects\TriNetX Code\Hypercapnia TriNetX CSV Processing\Working\\" #location where to read and right from (faster = better if space allows)
print(working_dir)
#num_spreadsheets = 10
num_spreadsheets = 853
```

### #############################################################################################

### Reformatting of CSVs

```python
# Adding new indexes; adding new column names; removing unneeded columns; reformatting column data types

columns = ["patient_id","encounter_id","code_system","code","date","value","text_value","units_of_measure","derived_by_TriNetX","source_id"]
for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    vital_signs = pd.read_csv(database_dir + "vital_signs"+f'{i:04}'+".csv",
                            names = columns,
                            dtype = {"patient_id":str,"encounter_id":str,"code":str},
                            parse_dates = ["date"])
    vital_signs = vital_signs.drop(columns = ["code_system","text_value","units_of_measure","derived_by_TriNetX","source_id"])
    vital_signs.to_csv(working_dir + "vital_signs_NEW_"+f'{i:04}'+".csv", index = False)
    print(vital_signs.shape)
    del vital_signs
```

### #############################################################################################

# Venous Blood Temperature; Processing
### LOINC: 75987-8 - Venous Blood Temperature
For some reason, TriNetX has these in vitals, not labs

```python
# LOINC: 75987-8 - Venous Blood Temperature

sub_blocks = []
for i in range(1, num_spreadsheets+1):
    print(f'{i:04}')
    vital_signs = pd.read_csv(working_dir + "vital_signs_NEW_"+f'{i:04}'+".csv", dtype={"patient_id":str, "encounter_id":str, "code":'category', "date":'category'})
    vital_signs = vital_signs[vital_signs["code"].str.match("^75987-8$")]
    vital_signs['value'] = vital_signs['value'].astype('float32') 
    vital_signs["value"] = np.where(vital_signs["value"] > 43.3, vital_signs["value"].apply(lambda x: (x-32)*(5/9)), vital_signs["value"]) #converts F to C if out of plausible range
    vital_signs = vital_signs.loc[(vital_signs['value'] < 43) & (vital_signs['value'] >= 20)]
    sub_blocks.append(vital_signs)
    del vital_signs

value_759878 = pd.concat(sub_blocks, ignore_index = True)
value_759878.to_csv(working_dir +"value_759878.csv", index = False)
display(value_759878.head())
print("\n")
print(value_759878.shape)
print("\n")
print(value_759878["code"].value_counts())
print("\n")
print("Number of NA/NaN Values:", value_759878.isna().sum().sum())

##### Histogram #####
plt.figure(figsize = (10,10))
plt.hist(value_759878["value"], bins = 500, facecolor = "blue")
plt.title("Histogram of value_759878", color = "black", fontsize = 12)
plt.xlabel("Value", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.show()
del value_759878
```
